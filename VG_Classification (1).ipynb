## Splitting data and preprocessing

```python
X = df.drop(['satisfaction'],axis=1)
y = df['satisfaction']
```

```python
X.nunique()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
for column in X.columns:
        if X[column].dtype == 'object':
            X[column] = label_encoder.fit_transform(df[column])
```

```python
X.head(3)
```

```python
y=label_encoder.fit_transform(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
y_train
```

## Model 1 using logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn import preprocessing
```

```python
model_1 = LogisticRegression(max_iter=100)
regression = model_1.fit(X_train, y_train)
```

```python
y_pred_1 = model_1.predict(X_test)
```

```python
regression.score(X,y)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(y_pred_1,y_test)
print(f"Accuracy: {accuracy:.2f} ")
```

## Model 2 using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model_2 = KNeighborsClassifier()
```

```python
model_2.fit(X_train, y_train)
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(y_pred_2,y_test)
print(f"Accuracy: {accuracy:.2f} ")
```

## Model 3 using Naive Bayes classifier

```python
from sklearn.naive_bayes import GaussianNB
model_3 = GaussianNB()
```

```python
model_3.fit(X_train, y_train)
```

```python
y_pred_3 = model_3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(y_pred_3,y_test)
print(f"Accuracy: {accuracy:.2f} ")
```

## Model 4 using random forest

```python
from sklearn.ensemble import RandomForestClassifier
model_4 = RandomForestClassifier()
```

```python
model_4.fit(X_train, y_train)
```

```python
y_pred_4= model_4.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy = accuracy_score(y_pred_3,y_test)
print(f"Accuracy: {accuracy:.2f} ")
```
